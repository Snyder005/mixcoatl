```python
%matplotlib inline
from astropy.io import fits
from mixcoatl.sourcegrid import *
import matplotlib.pyplot as plt
h = fits.open(f)
grid = DistortedGrid(h[1].header['YSTEP'],h[1].header['XSTEP'], h[1].header['THETA'], h[1].header['Y0'], h[1].header['X0'], 49, 49)
grid.make_source_grid()
gY,gX = grid.get_source_centroids()

fig, ax = plt.subplots(1, 1, figsize=(11, 10))
ax.scatter(gY, gX, color='blue', marker='x')
ax.set_xlim(0,4000)
ax.set_ylim(0,4000)
```
